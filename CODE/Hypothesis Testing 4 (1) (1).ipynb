```python
import pandas as pd
import numpy as np

df1 = pd.read_csv("C:/Users/miche/Documents/datasets/countries_clean5.csv");
df1
```

```python
data_country = df1.groupby(by=['country','year','HDI_for_year','gdp_per_capita']).agg({'suicide_no': ['sum'],'population': ['sum']})
data_country.columns = ['suicide_no','population']
data_country.reset_index(inplace=True)
data_country
```

```python
data_country["suicides_per_100k"] = data_country["suicide_no"] * 100000/data_country["population"]
data_country
```

```python
new_df_2003 = data_country.loc[(data_country["year"] == 2003)]
new_df_2003.reset_index(drop=True, inplace=True)
new_df_2003
#new_df_2003=new_df_2003.drop('index')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
def diagnostic_plots(countries, variable):
    # plot histogram and Q-Q plot side by side
    plt.figure(figsize=(15,6))
    plt.subplot(1,2,1)
    countries[variable].hist()
    
    plt.subplot(1,2,2)
    stats.probplot(countries[variable], dist="norm", plot=plt)
    
    plt.show()
    
```

```python
new_df_2003['suicides_per_100k_log'] = np.log(new_df_2003['suicides_per_100k'])
diagnostic_plots(new_df_2003,"suicides_per_100k_log")
```

```python
new_df_2003.describe()
```

```python
sample2=new_df_2003.loc[new_df_2003['HDI_for_year'] >0.801325]
sample2.reset_index(drop=True, inplace=True)
sample2
```

```python
sample2.describe()
```

```python
np.random.seed(2)
test_sample = sample2.sample(frac =.5) 
test_sample
```

```python
test_sample.describe()
```

```python
#USING NORMALISED DATA
#Does HDI have an effect on suicies_per_100k?

#PROBLEM STATEMENT
#A set of countries whose HDI was greater than 0.801325 wanted to prove that their mean suicides_per_100k in 2003 
#would be lesser than the population mean suicides_per_100k
#H0:The total suicides_per_100k is more than or equal to 12.588647
#H1:The total suicides_per_100k is less than 12.588647(log value: 2.368484)

tst, pval=stats.ttest_1samp(test_sample["suicides_per_100k_log"],               # Sample data
                 2.368484)
print("tst",tst)
print("pvalue",pval/2)
if pval/2 < 0.05:    # alpha value is 0.05 or 5%
   print(" we are rejecting null hypothesis")
else:
  print("we cannot reject null hypothesis")
```

```python
#H0 is plausible
#Conclusion: There is not sufficient evidence to support the claim that the total_suicides_per_100k 
#will be lesser than 12.588647 if the HDI is greater than 0.801325
```

```python
sample_gdb=new_df_2003.loc[new_df_2003['gdp_per_capita'] >21346.625]
sample_gdb.reset_index(drop=True, inplace=True)
sample_gdb
```

```python
np.random.seed(3)
test_sample_gdb = sample_gdb.sample(frac =.65) 
test_sample_gdb
```

```python
#USING NORMALISED DATA
#Does GDP have an effect on suicies_per_100k?


#A set of countries whose GDP was greater than 21346.625 wanted to prove that their mean suicides_per_100k in 2003 
#would be lesser than the population mean suicides_per_100k
#H0:The total suicides_per_100k is more than or equal to 12.588647
#H1:The total suicides_per_100k is less than 12.588647
tset, pval=stats.ttest_1samp(test_sample_gdb["suicides_per_100k_log"],               # Sample data
                 2.368484)
print("tst",tset)
print(pval/2)
if pval/2 < 0.05:    # alpha value is 0.05 or 5%
   print(" we are rejecting null hypothesis")
else:
  print("we cannot reject null hypothesis")
```

```python

#We reject H0 as p<0.05 and accept alternate hypothesis
#Our results are statistically significant at 5% level

#Conclusion: There is some evidence that total_suicides_per_100k will be lesser than
#12.588647 if the GDP is greater than 21346.625 
```
