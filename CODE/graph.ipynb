```python
import pandas as pd
import numpy as np

df = pd.read_csv("countries_clean5.csv");
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df
```

```python
sns.set(style='whitegrid')
sns.boxplot(df.HDI_for_year)
plt.show()
```

```python
data_year_no = df.groupby(by=['year', 'sex']).agg({'suicide_no': ['sum']})
data_year_no.columns = ['suicide_no']
data_year_no.reset_index(inplace=True)
data_year_no

color = sns.color_palette("Set2")
sns.set(style="darkgrid")
plt.figure(figsize=(15, 10))
sns.barplot(x='year', y='suicide_no', hue='sex', data=data_year_no, palette=color)
plt.xticks(rotation=90)

plt.figtext(.5,.91,'Worldwide Suicide by Year (1990 - 2013)', color='b', fontsize=25, ha='center')
plt.show()
```

```python
data_year_no = df.groupby(by=['year']).agg({'suicide_no': ['sum']})
data_year_no.columns = ['suicide_no']
data_year_no.reset_index(inplace=True)
data_year_no

color = sns.color_palette("Set2")
sns.set(style="darkgrid")
plt.figure(figsize=(15, 10))
sns.barplot(x='year', y='suicide_no', data=data_year_no)
plt.xticks(rotation=90)

plt.figtext(.5,.91,'Worldwide Suicide by Year (1990 - 2013)', color='b', fontsize=25, ha='center')
plt.show()

#we can conclude that the highest no of suicides were in the year of 2003
```

```python
data_country_no = df.groupby(by=['country']).agg({'suicide_no': ['sum']})
data_country_no.columns = ['suicide_no']
data_country_no.reset_index(inplace=True)
data_year_no

color = sns.color_palette("Set2")
sns.set(style="darkgrid")
plt.figure(figsize=(15, 10))
sns.barplot(x='country', y='suicide_no', data=data_country_no)
plt.xticks(rotation=90)

plt.figtext(.5,.91,'Worldwide Suicide by Year (1990 - 2013)', color='b', fontsize=25, ha='center')
plt.show()
#it is evident from our dataset that Russia has the most number of suicides
```
